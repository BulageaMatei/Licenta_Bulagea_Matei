```python
#Pregătim dataseturile pentru a le putea utiliza

transform = transforms.Compose(
[transforms.Resize((224,224)),                           # redimensionăm imaginiile să fie de 224 x 224 
 transforms.RandomAffine(0, shear=10, scale=(0.8,1.2)),
 transforms.RandomHorizontalFlip(p = 0.2),             # să rotim imagiinle random cu p = 0.2(sa învețe din mai multe unghiuri)
 transforms.RandomVerticalFlip(p = 0.2),
 transforms.ToTensor(),                                # să îl facă tensor
 transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])])   # la fel, luate pentru modelul de ResNet50 : mean, std 


transform_test = transforms.Compose(
[transforms.Resize((224,224)),
 transforms.ToTensor(),
 transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])])

###################
# 1. Identifcare masca cu ResNet-50

#Importam datasets pentru antrenare, validare si testare ( din retea locala)
image_datasets = {
    'train':
     datasets.ImageFolder("E:\\Profil\\Desktop\\Licenta\\ISIA\\LICENTA\\12k\\Face_Mask_Dataset\\Train", transform = transform),
    'validation':
     datasets.ImageFolder("E:\\Profil\\Desktop\\Licenta\\ISIA\\LICENTA\\12k\\Face_Mask_Dataset\\Validation", transform = transform),
    'test':
     datasets.ImageFolder("E:\\Profil\\Desktop\\Licenta\\ISIA\\LICENTA\\12k\\Face_Mask_Dataset\\Test", transform = transform_test)
}

#Cream data loaders pentru datasets; shuffle pentru antrenare doar
dataloaders = {
    'train':
     torch.utils.data.DataLoader(image_datasets['train'], batch_size = 16, shuffle = True, num_workers = 4),
    'validation':
     torch.utils.data.DataLoader(image_datasets['validation'], batch_size = 16, shuffle = True, num_workers = 4),
    'test':
     torch.utils.data.DataLoader(image_datasets['test'], batch_size = 1, shuffle = False, num_workers = 4)
}

#Verificam dimensiunile
print('Training set has {} instances'.format(len(image_datasets['train'])))
print('Validation set has {} instances'.format(len(image_datasets['validation'])))
print('Test set has {} instances'.format(len(image_datasets['test'])))
```

```python
if torch.cuda.is_available():
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
print(device)
```

```python
from torchvision.models import resnet50, ResNet50_Weights

#model = models.resnet50(weights=ResNet50_Weights.DEFAULT).to(device) # cele mai actuale weights, dar are probleme 
model = models.resnet50(pretrained=True).to(device)   #importam resnet


# freezing pretrained layers
for param in model.parameters():
    param.requires_grad = True 

model.fc = nn.Sequential(
               nn.Dropout(p = 0.5, inplace = False),
               nn.Linear(in_features = 2048, out_features =128, bias = True),
               nn.ReLU(inplace=True),
               nn.Dropout(p = 0.5, inplace = False),
               nn.Linear(in_features = 128, out_features = 2, bias = True)).to(device)
```

```python
model
```

```python
loss_f = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr = 1e-4)
```

```python
global_train_loss = []
global_validation_loss = []    #sa se vada frumos pe grafice, cu toate ca nu e
val_accuracy = []                  #nevoie din moment ce reteaua e preantrenata
train_accuracy = []
```

```python
def train_model(model, criterion, optimizer, num_epochs):
    
    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch+1, num_epochs))
        print('-' * 10)
         
        
        for phase in (['train', 'validation']):
            if phase == 'train':
                model.train()
            else:
                model.eval()    #switch pt layerele/partile din model care se comporta diferit in timpul antrenarii/evaluarii

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)     #transmitem imaginile / labels catre dispozitivul curent unde se afla modelul
                labels = labels.to(device)

                outputs = model(inputs)        # transmitem imaginile in modelul nostru si sa le vedem pe output
                
                loss = loss_f(outputs, labels)  #

                if phase == 'train':
                    optimizer.zero_grad()  # sa imi faca gradientii 0
                    loss.backward()        # sa computeze gradientii      # pe scurt : backprop
                    optimizer.step()       #update weights utilizand gradientii

                _, preds = torch.max(outputs, 1)
                running_loss += loss.detach() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / len(image_datasets[phase])
            epoch_acc = running_corrects.float() / len(image_datasets[phase])

            print('{} loss: {:.7f}, acc: {:.7f}'.format(phase,                         #sa afisam cu 4 zecimale {:4f}
                                                        epoch_loss.item(),             #contine loss-value pt o epoca (tensor)
                                                        epoch_acc.item()))             #contine acuratetea pt o epoca (tensor)
            if phase =='train':
                global_train_loss.append(epoch_loss.item())
                train_accuracy.append(epoch_acc.item())
            if phase == 'validation':
                global_validation_loss.append(epoch_loss.item())
                val_accuracy.append(epoch_acc.item())
            
    return model, global_train_loss, train_accuracy, global_validation_loss, val_accuracy
 
```

```python
model_trained, *_ = train_model(model, loss_f, optimizer, num_epochs=10) 
```

```python
#Definim calea ca sa salvam modelul
model_path = 'buli_nr_1.pth'

#Salvam modelul
torch.save(model_trained.state_dict(), model_path)
```

```python
plt.plot(global_train_loss, marker = 'o')
plt.plot(train_accuracy, marker = 'o')

plt.plot(global_validation_loss, marker = 'o')
plt.plot(val_accuracy, marker = 'o')


plt.xlabel("Epochs")
plt.ylabel("Loss/Accuracy")
plt.grid()

plt.legend(["Training Loss", "train_accuracy", "Validation Loss", "val_accuracy"])
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
import pandas as pd

y_pred = []
y_true = []

for inputs, labels in dataloaders['test']:
    inputs = inputs.to(device) 
    labels = labels.to(device)
    
    output = model(inputs)
    
    #output = F.softmax(output, dim=1).cpu().data.numpy()
    output = F.sigmoid(output).cpu().data.numpy()
    y_pred.extend(output) #Save Predictions
    
    labels = labels.data.cpu().numpy()
    y_true.extend(labels) # Save Truth
    
classes = ('Mască', 'No_Mască')

y_pred = np.argmax(y_pred, axis=1)

cf_matrix = confusion_matrix(y_true, y_pred)

df_cm = pd.DataFrame(cf_matrix, index=[i for i in classes], columns=[i for i in classes])

ax = plt.gca()
plt.figure(figsize=(12, 7))


sn.heatmap(df_cm, annot=True, fmt='d', cmap='Blues', cbar=True, ax=ax)

ax.set_xlabel('Predicție')
ax.set_ylabel('Actual')
ax.set_title('Matrice de confuzie')

plt.xticks(rotation=45)
plt.yticks(rotation=0)

#plt.savefig('output.png')
```
